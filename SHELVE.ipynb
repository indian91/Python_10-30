```python
from getpass import getpass
def signup():
    db=shelve.open('BankApp/data.db',writeback=True)
    d={'Name':input('Enter yoyr Name'),'Bal':int(input('Enter Balance Amount')),'Pwd':getpass('Enter Password')}
    acc=db['last_acc']
    acc+=1
    db[str(acc)]=d
    db['last_acc']=acc
    db.close()
```
